```python
import pandas as pd
```

```python
# dataset is taken from Kaggle website (https://www.kaggle.com)
df = pd.read_csv('./datasets/titanic.csv', index_col='PassengerId')
```

```python
df.head()
```

```python
df.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc='mean')
```

```python
df.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc='mean', margins=True)
```

```python
df.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc='count', margins=True)
```
