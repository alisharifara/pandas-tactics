```python
import pandas as pd
import numpy as np
# dataset is taken from Kaggle website (https://www.kaggle.com/saurabh00007/diabetescsv)
```

```python
df = pd.read_csv('./datasets/diabetes.csv')
```

```python
df.head()
```

```python
# Add summary rows 
df.loc['ROW_SUM'] = df.apply(lambda x: x.sum())
df.loc['ROW_AVG'] = df.apply(lambda x: x.mean())
df.loc['ROW_MAX'] = df.apply(lambda x: x.max())
df.loc['ROW_MIN'] = df.apply(lambda x: x.min())
```
