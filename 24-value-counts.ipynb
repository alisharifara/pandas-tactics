```python
# This is a function to check value distributions
# some arguments for value_counts:
# 1. normalize = True: if you want to check the frequency instead of counts.
# 2. dropna = False: if you also want to include missing values in the result.

import pandas as pd
```

```python
df = pd.read_csv('./datasets/titanic.csv', index_col='PassengerId')
```

```python
df.head()
```

```python
df['Sex'].value_counts(dropna=False).reset_index().sort_values(by='index')
```

```python
df['Age'].value_counts(dropna = False)
```

```python
df['Age'].value_counts(normalize = True)
```

```python
df['Cabin'].value_counts(dropna = False)
```
