```python
import pandas as pd
import numpy as np

# in SQL it's equivalent to:
# SELECT Sex, Pclass, max(Age), avg(Age), sum(Fare), min(Fare), count(*) FROM table GROUP BY Sex, Pclass
# in pandas, we can use groupby() with agg()
# dataset is taken from the Kaggle website (https://www.kaggle.com/heptapod/titanic)
```

```python
df = pd.read_csv('./datasets/titanic.csv', index_col='PassengerId')
```

```python
df.head()
```

```python
results = df.groupby(['Sex','Pclass']).agg(
                                {'Age':  ['max', np.mean],
                                 'Fare': ['sum','min','count']
                                })
```
