if you wanna filter values of a column based on conditions from another set of columns you can use the following code:


```python
import pandas as pd
```

```python
# dataset is taken from Kaggle website (https://www.kaggle.com)
df = pd.read_csv('./datasets/titanic.csv')
```

```python
df.head(10)
```

```python
# it returns the rows which the passengers were female and their ages are less than 15 and their fares were less than $100 and survived!!!  
df.loc[(df['Sex']== 'female') &
       (df['Age'] < 15) &
       (df['Fare'] < 100) &
       (df['Survived']== 1) , ['Name', 'Age', 'Ticket','Fare']]
```
